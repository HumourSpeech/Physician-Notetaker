Converting labels..

```python
def convert_file(input_path, output_path, label_map):
    with open(input_path, "r", encoding="utf-8") as f_in, \
         open(output_path, "w", encoding="utf-8") as f_out:

        for line in f_in:
            line = line.strip()

            # Preserve sentence boundaries
            if not line:
                f_out.write("\n")
                continue

            parts = line.split()
            if len(parts) != 2:
                # Safety fallback
                f_out.write(line + "\n")
                continue

            token, label = parts
            new_label = label_map.get(label, "O")

            f_out.write(f"{token} {new_label}\n")


# ---- Run conversion ----
LABEL_MAP = {
    "B-problem": "B-SYMPTOM",
    "I-problem": "I-SYMPTOM",
    "B-treatment": "B-TREATMENT",
    "I-treatment": "I-TREATMENT",
    "B-test": "O",
    "I-test": "O",
    "O": "O"
}

```

```python
convert_file("../Data/i2b2/train.txt", "../Data/i2b2_Transformed/train.txt", LABEL_MAP)
convert_file("../Data/i2b2/dev.txt", "../Data/i2b2_Transformed/dev.txt", LABEL_MAP)
convert_file("../Data/i2b2/test.txt", "../Data/i2b2_Transformed/test.txt", LABEL_MAP)
```
